```python
# 13. Model tính tổng hai tensor
class SumModel(tf.Module):
    @tf.function
    def __call__(self, x, y):
        return x + y

sum_model = SumModel()
print("\n13. Kết quả SumModel:\n", sum_model(x12, y12).numpy())
```

```python
# 14. Model tính hiệu và tích
class DiffProdModel(tf.Module):
    @tf.function
    def compute(self, x, y):
        return x - y, x * y

diff_prod_model = DiffProdModel()
diff, prod = diff_prod_model.compute(x12, y12)
print("\n14. Kết quả DiffProdModel (Hiệu):\n", diff.numpy())
```

```python
# 15. Model tính chu vi hình chữ nhật P = (a + b) * 2
class PerimeterModel(tf.Module):
    @tf.function
    def calculate(self, a, b):
        return (a + b) * 2

peri_model = PerimeterModel()
print("\n15. Chu vi hcn (a=3, b=4):", peri_model.calculate(3.0, 4.0).numpy())
```

```python
# 16. Model tính diện tích toàn phần hình hộp chữ nhật
class SurfaceAreaModel(tf.Module):
    @tf.function
    def calculate(self, a, b, c):
        s_xung_quanh = (a + b) * c * 2
        s_toan_phan = (a * b) * 2 + s_xung_quanh
        return s_toan_phan

sa_model = SurfaceAreaModel()
print("\n16. Diện tích toàn phần (a=2, b=3, c=4):", sa_model.calculate(2.0, 3.0, 4.0).numpy())
```

```python
# Bài 17: Chứng minh AD // BC (Trường hợp True)
import tensorflow as tf

class ProveParallelModel(tf.Module):
    @tf.function
    def _is_equal_length(self, u, v):
        len_u_sq = tf.reduce_sum(tf.square(u))
        len_v_sq = tf.reduce_sum(tf.square(v))
        return tf.abs(len_u_sq - len_v_sq) < 1e-5

    @tf.function
    def prove_bai17(self, A, B, C, D):
        AB = B - A
        CD = D - C
        AD = D - A
        BC = C - B

        cond_AD_equal_BC = self._is_equal_length(AD, BC)
        cond_AB_equal_CD = self._is_equal_length(AB, CD)
        return cond_AD_equal_BC & cond_AB_equal_CD

# Tọa độ giả lập để kết quả ra True (Hình thang cân ABCD)
A = tf.constant([0.0, 2.0])
B = tf.constant([2.0, 2.0])
C = tf.constant([3.0, 0.0])
D = tf.constant([1.0, 0.0])

model_17 = ProveParallelModel()
print(f"Kết quả Bài 17 (AD // BC): {model_17.prove_bai17(A, B, C, D).numpy()}")
```

```python
# Bài 19: Định lý đường trung bình (Sửa tọa độ để MN = MB ra True)
import tensorflow as tf

class ProveMidpointModel(tf.Module):
    @tf.function
    def _is_parallel(self, u, v):
        return tf.abs(u[0] * v[1] - u[1] * v[0]) < 1e-5

    @tf.function
    def _is_equal_length(self, u, v):
        return tf.abs(tf.reduce_sum(tf.square(u)) - tf.reduce_sum(tf.square(v))) < 1e-5

    @tf.function
    def prove_bai19(self, M, N, B, C):
        MN = N - M; BC = C - B; MB = B - M
        return self._is_equal_length(MN, MB) & self._is_parallel(MN, BC)

M = tf.constant([0.0, 2.0])
N = tf.constant([2.0, 1.0])
B = tf.constant([-1.0, 0.0])
C = tf.constant([3.0, -2.0])

model_19 = ProveMidpointModel()
print(f"Kết quả Bài 19 (NA = NC): {model_19.prove_bai19(M, N, B, C).numpy()}")
```
